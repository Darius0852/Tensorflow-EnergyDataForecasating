Import Libraries

```
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from matplotlib.pyplot import figure




```

Import dataset and split into matrix of features and matrix of dependatn variables

```
dataset = pd.read_csv("house1_energy_blk1.csv")
print(dataset)

#Sub Energy Data into DataFrame
energyBathroomsOrig = dataset.sub9
energyBathrooms = pd.DataFrame(energyBathroomsOrig)
print(energyBathrooms)

#Time of Day values in corresponding DataFrame
timeUnix = dataset.unix_ts
print(timeUnix)

#change each unix time into normal time 
timeReadArray = []

for x in timeUnix:
  try:
    x = int(x)
    timeReadable = dt.datetime.fromtimestamp(x)
    timeReadArray.append(timeReadable)
  except:
    print("An exception occurred")

print(timeReadArray)
df = pd.DataFrame(timeReadArray)
print(df)


```

Produce the Graph of energy over time

```
#plot points of bathroom per hour Feb 7th 8am to 16th Feb 7am
plt.style.use('seaborn-whitegrid')

figure(num=None, figsize=(20, 16), dpi=120, facecolor='w', edgecolor='k')
plt.plot(timeUnix, energyBathroomsOrig)
plt.xlabel('Time (S)')
plt.ylabel('Energy (kW)')
```
